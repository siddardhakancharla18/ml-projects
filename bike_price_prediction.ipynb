```python
import seaborn as sns
```

```python
sns.boxplot(data)
```

```python
from scipy import stats
import numpy as np

# Compute Z-scores
z_scores = np.abs(stats.zscore(data[['km_driven','ex_showroom_price','price']]))

# Set a threshold (commonly 3)
data = data[(z_scores < 3).all(axis=1)]
```

```python
data = data[(z_scores < 3).all(axis=1)]
```

```python
data
```

```python
sns.boxplot(data)
```

```python
data = data.drop(columns = ['brand_encoded'])
```

```python
x = data.drop(['price'],axis = 1)
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)

rf = RandomForestRegressor()
rf.fit(x_train,y_train)

gb = GradientBoostingRegressor()
gb.fit(x_train,y_train)

xgb = XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
y_pred1 = lr.predict(x_test)
y_pred2 = rf.predict(x_test)
y_pred3 = gb.predict(x_test)
y_pred4 = xgb.predict(x_test)
```

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
#visualizing the best model among
final_data = pd.DataFrame({'models':['lr','rf','gb','xgb'],
            "r2score":[score1,score2,score3,score4]})
```

```python
final_data
```

```python
sns.barplot(x=final_data['models'], y=final_data['r2score'])
```

```python
#applying xgb on entire dataset
gb = XGBRegressor()
gb_final = gb.fit(x,y)
```

```python
#saving the model
import joblib
```

```python
joblib.dump(gb_final,'bike_price_predictor')
```
